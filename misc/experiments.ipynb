```python
import os

import whisper
from IPython.display import Audio
```

```python
yt_data_path = r"F:/BIG_FILES/AI_DATA/2024_STT"
all_vids = os.listdir(yt_data_path)
vid = all_vids[3]
vid_path = os.path.join(yt_data_path, vid)
vid_segments = os.listdir(os.path.join(vid_path, "segments"))
vid_transcripts = os.listdir(os.path.join(vid_path, "transcripts"))

segment = vid_segments[0]
segment_path = os.path.join(vid_path, "segments", segment)
transcript_path = os.path.join(vid_path, "transcripts", vid_transcripts[0])

transcript = open(transcript_path, "r", encoding="utf-8").read()
```

```python
model = whisper.load_model("medium", download_root="./cache", device="cuda")
```

```python
result = model.transcribe(segment_path, language="de")
```

```python
display(Audio(segment_path))
print("Target:", transcript.strip())
print("Result:", result["text"].strip())
```
